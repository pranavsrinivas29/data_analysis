```python
#Using Apply function
def triple_age(x):
  return x*3

df['Tripled_Age'] = df['Age'].apply(triple_age)
df.head(2)
```

```python
# Creating a dictionary to map values
# Used less values in mapping for easy execution
age_mapping = {25.0: 'Middle-Age', 14.0: 'Young-aged', 19.0: 'Adult'}

# Creating a new column 'Age_Category' using map
df['Age_Category'] = df['Age'].map(age_mapping)

print(df.head(2))
```

```python
# Creating map for all values
# Mapping age ranges using a dictionary
# Converting float values to integers
# Handling NaN values and converting the 'Age' column to integers
df['Age'] = df['Age'].fillna(0).astype(int)


df.head(5)


# Mapping age ranges using a dictionary
age_ranges = {range(0, 30): 'Young', range(30, 40): 'Middle-aged', range(40, 100): 'Old'}

# Applying the mapping using map
df['Age_Category'] = df['Age'].map(lambda x: next((v for k, v in age_ranges.items() if x in k), None))

df.head(10)
```

```python
# Merge
# Creating two DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Salary': [50000, 60000, 70000]})

# Merging on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')

print(merged_df)
```

```python
# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
        'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
        'Score': [85, 90, 88, 92, 89, 94]}
df = pd.DataFrame(data)

# Grouping by 'Name' and calculating the mean score
grouped_data = df.groupby('Name')['Score'].mean().reset_index()

print(grouped_data)
```

```python
# Grouping by 'Name' and calculating the total score
grouped_data = df.groupby('Name')['Score'].sum().reset_index()

print(grouped_data)
print('')
# Grouping by 'Name' and calculating the subject count
grouped_data = df.groupby('Name')['Subject'].count().reset_index()

print(grouped_data)
```
