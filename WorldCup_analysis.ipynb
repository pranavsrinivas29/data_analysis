```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_cup.csv')  
```

```python
df
```

```python
df['Host_Champions']='No'
df.loc[df['Champion'] == df['Host'], 'Host_Champions'] = 'Yes'
df
```

```python
# Count each category
counts = df['Host_Champions'].value_counts()
print(counts)
```

```python
# Plot the bar plot
plt.figure(figsize=(8, 6))
plt.bar(counts.index, counts.values, color=[ 'red','green'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Counts of Host being Champions')
plt.show()
```

```python
# Plot the line plot
plt.figure(figsize=(20, 10))
#plt.boxplot(df['Year'], df['AttendanceAvg'])
sns.boxplot( data=df, x='Year', y='AttendanceAvg',hue='Host_Champions',  palette='Set3',width=.5,color=".8", linewidth=.75)
#sns.stripplot(x='Year', y='AttendanceAvg', data=df, color='black', alpha=0.5, jitter=True)
plt.xlabel('Year')
plt.ylabel('Average Attendance')
plt.title('Average Attendance Over Years')
plt.grid(True)
plt.show()
```

```python
df['Host_Runners']='No'
df.loc[df['Runner-Up'] == df['Host'], 'Host_Runners'] = 'Yes'
df
```

```python
Host_Runners_count, Host_Champions_count = df[df['Host_Runners']=='Yes'].shape[0], df[df['Host_Champions']=='Yes'].shape[0]
# Prepare data for bar plot
labels = ['Host Runners', 'Host Champions']
counts = [Host_Runners_count, Host_Champions_count]

plt.figure(figsize=(5,5))
plt.bar(labels, counts, color='skyblue')
```

```python
# Filter the DataFrame for host champions and host runners
host_champions = df[df['Host_Champions'] == 'Yes']
host_runners = df[df['Host_Runners'] == 'Yes']
# Extract unique country names of host champions
host_champion_countries = host_champions['Host'].unique()
# Extract unique country names of host runners
host_runner_countries = host_runners['Host'].unique()

print("Countries who are host champions:")
print(host_champion_countries)

print("Countries who are host runners:")
print(host_runner_countries)
```

```python
winners=df['Champion'].value_counts()


runners=df['Runner-Up'].value_counts()
runners
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))

# Plot the count of countries being champions
axes[0].bar(winners.index, winners.values, color='skyblue')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Count')
axes[0].set_title('Count of Countries Being Champions')
axes[0].tick_params(axis='x', rotation=90)

# Plot the count of host countries who were champions
axes[1].bar(runners.index, runners.values, color='lightgreen')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Count')
axes[1].set_title('Count of Countries Being Runners')
axes[1].tick_params(axis='x', rotation=90)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Get the unique values of champions and runners-up
champions = set(df['Champion'])
runners_up = set(df['Runner-Up'])

# Find countries that have been both champions and runners-up
both_winners_and_runners = champions.intersection(runners_up)

print("Countries that have been both champions and runners-up:")
print(both_winners_and_runners)
```

```python
# Align the indices of both series
combined_counts = winners.to_frame('Champion').join(runners.to_frame('Runner-Up'), how='outer').fillna(0)
print(combined_counts)
# Find countries that have the same number of championships and runner-up finishes
same_counts = combined_counts[combined_counts['Champion'] == combined_counts['Runner-Up']]

# Print the results
print("Countries that have been champions and runners-up the same number of times:")
print(same_counts)
```
