Modifying Data within DataFrames

```python
df.columns=[x.upper() for x in df.columns]
df.head(10)
```

```python
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
df.rename(columns={'EMPLOYMENT':'EMP','HOBBYIST':'HOBBY'},inplace=True)
df.head(5)
```

```python
df.at[1,'OPENSOURCER']='ALWAYS'
df.head(5)
```

```python
def updatecountry(STUDENT):
  return STUDENT.upper()

df['STUDENT'].apply(updatecountry)
```

```python
df['STUDENT'] = df['STUDENT'].apply(lambda x:x.lower())
```
